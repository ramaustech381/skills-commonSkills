```python
#Covert .json annotated data into .spacy format
import srsly
import typer
import warnings
from pathlib import Path

import spacy
from spacy.tokens import DocBin


def convert(lang: str, input_path: Path, output_path: Path):
    nlp = spacy.blank(lang)
    db = DocBin()
    for annotated in srsly.read_json(input_path):
        doc = nlp.make_doc(annotated["content "])
        ents = []
        ll = False
        for entities in annotated["annotation"]:
          for point in entities['points']:
            span = doc.char_span(point['start'], point['end'], label=entities['label'])
            if span is None:
                ll = True
                break
                # msg = f"Skipping entity [{point['start']}, {point['end']}, {entities['label']}] in the following text because the character span '{doc.text[point['start']:point['end']]}' does not align with token boundaries:\n\n{repr(annotated['content'])}\n"
                # warnings.warn(msg)
            else:
                ents.append(span)
          if ll == True:
            break
        doc.ents = ents
        print(doc.ents)
        db.add(doc)
    db.to_disk(output_path)


# if __name__ == "__main__":
#     typer.run(convert)
```
